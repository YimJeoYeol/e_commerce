```python
# 레이블을 특정 문자열로 일괄변환하는 함수 생성
def 변환(x):
    clus_dict = {0: "20_중소득_중간점",
                 1: "40_고소득_최저점",
                 2: "50_중소득_중간점",
                 3: "30_고소득_최고점",
                 4: "20_저소득_최고점",
                 5: "40_저소득_최저점"}
    if x == 0:
        return clus_dict[0]
    elif x == 1:
        return clus_dict[1]
    elif x == 2:
        return clus_dict[2]
    elif x == 3:
        return clus_dict[3]
    elif x == 4:
        return clus_dict[4]
    elif x == 5:
        return clus_dict[5]
    else:
        pass

data.groupby("label").mean()
```

```python
# "변환"함수 적용
data["label"] = data["label"].apply(변환)
```

# 주성분 분석

```python
# 주성분 분석을 위해 임포트
from sklearn.decomposition import PCA
```

```python
# PCA 모델 생성(컴포넌트 갯수는 2)
pca = PCA(n_components=2)
```

```python
pca.fit(df)
```

```python
# 주성분 분석을 통해 차원 축소
df_pca = pca.transform(df)
```

```python
df_pca
```

```python
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"])
```

```python
df_pca["label"] = data["label"]
```

```python
# 시각화
plt.figure(figsize=(20,10))
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["label"], palette="Set2", s=100)
```
