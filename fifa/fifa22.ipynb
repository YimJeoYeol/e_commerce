```python
# 나라별로 선수들의 wage 총합이 가장 큰 나라 10개
# groupby("nationality")
국가별_주급 = df[["wage_eur", "nationality"]].groupby("nationality")
국가별_주급.sum().nlargest(10, "wage_eur")
```

```python
df.pivot_table("wage_eur", index="nationality", aggfunc=sum).sort_values("wage_eur", ascending=False)
```

```python
# 대한민국의 wage 총합 랭킹은 세계 몇위일까?
df_wage = df.pivot_table("wage_eur", index="nationality", aggfunc=sum).sort_values("wage_eur", ascending=False).rank(ascending=False)
```

```python
# 나라별로 평균임금과 표준편차
df_wage.describe()
```

```python
국가별_주급 = df[["wage_eur", "nationality"]].groupby("nationality")
국가별_주급
```

```python
# 국가별 주급 평균이 가장 높은 나라 순으로 정렬
국가별_주급.describe()["wage_eur"][["mean", "std"]].sort_values("std", ascending=False)
```

```python
# "국가별_주급" 데이터를 활용해서, 대한민국의 평균주급은 전세계 몇위인지?
국가별_주급.describe()["wage_eur"]["mean"].rank(ascending=False).loc[["Korea Republic", "England", "Germany", "Brazil", "France", "Italy"]]
```
